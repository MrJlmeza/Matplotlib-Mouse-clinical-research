```python
#IMPORT DEPENDENCIES
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as sts

#FILES TO LOAD
mouse_metadata_path = "./data/Mouse_metadata.csv"
study_results_path = "./data/Study_results.csv"

#READ CSV FILES AS DATAFRAMES
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

#MERGE DATA FRAMES WITHOUT DUPLICATE MOUSE IDS
combined_df = pd.merge(mouse_metadata, study_results, how='outer', on='Mouse ID')
combined_df
```

```python
#CHECK THE NUMBER OF MICE
mice = combined_df["Mouse ID"].nunique()
mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint
duplicate_mice = combined_df.loc[combined_df.duplicated(subset=["Mouse ID", "Timepoint",]),"Mouse ID"].unique()
duplicate_mice_df = pd.DataFrame(duplicate_mice)
duplicate_mice_df
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
combined_clean_df = combined_df[combined_df['Mouse ID'].isin(duplicate_mice)==False]
combined_clean_df
```

```python
 # Checking the number of mice in the clean DataFrame.
combined_clean_df["Mouse ID"].nunique()
```

```python
volMean = combined_clean_df.groupby('Drug Regimen').mean()['Tumor Volume (mm3)']
volMed = combined_clean_df.groupby('Drug Regimen').median()['Tumor Volume (mm3)']
volStd = combined_clean_df.groupby('Drug Regimen').std()['Tumor Volume (mm3)']
volVar = combined_clean_df.groupby('Drug Regimen').var()['Tumor Volume (mm3)']
volSem = combined_clean_df.groupby('Drug Regimen').sem()['Tumor Volume (mm3)']
summ_stat_df = pd.DataFrame({"Mean": volMean, 
                             "Median": volMed,
                             "Var": volStd,
                             "std": volVar,
                             "sem": volSem})
summ_stat_df
```

```python
# Using the aggregation method, generate a summary statistics table in a single line for:
## [mean] [median] [variance] [standard deviation] [SEM]
drug_summary_df = combined_df.groupby("Drug Regimen").agg({"Tumor Volume (mm3)": ["mean", "median", "var", "std", "sem",],})
drug_summary_df
```

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
timepoint_df = combined_clean_df.groupby(["Mouse ID"])["Timepoint"].max()

timepoint_df = timepoint_df.reset_index()
# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
final_vol_df = pd.merge(timepoint_df, combined_clean_df, how="left", on=["Mouse ID", "Timepoint"])

final_vol_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumorVol = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for treatment in treatments:
    
   
    # Locate the rows which contain mice on each drug and get the tumor volumes
    treatments_df = final_vol_df.loc[final_vol_df["Drug Regimen"] == treatment]
    volume = treatments_df["Tumor Volume (mm3)"]
    
    # add subset 
    tumorVol.append(volume)
    
    # Determine outliers using upper and lower bounds
    quartiles = volume.quantile([.25, .5, .75])
    lowerq = quartiles[.25]
    upperq = quartiles[.75]
    iqr = upperq-lowerq
    
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    
    outliers = (volume.loc[
        (treatments_df["Tumor Volume (mm3)"] <= lower_bound) |
        (treatments_df["Tumor Volume (mm3)"] >= upper_bound)]).count()
    
   
    print(f"Drug: {treatment}")
    print(f"    IQR: {iqr}")
    print(f"    Upper Bound: {upper_bound}")
    print(f"    Lower Bound: {lower_bound}")
    print(f"    Number of Outliers: {outliers}")
    
```

```python
 # Generate a box plot of the final tumor volume of each mouse across four regimens of interest
plt.boxplot(tumorVol)

plt.title('Final Tumor Volume by Drug')
plt.ylabel('Final Tumor Volume (mm3)')
plt.xticks([1, 2, 3, 4], ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin'])
plt.show()
```

```python
# Generate a scatter plot of average tumor volume vs. mouse weight for the Capomulin regimen
capomulin_df = combined_clean_df.loc[combined_clean_df["Drug Regimen"] == "Capomulin"]
capomulin_df

capomulin_vol_df = pd.DataFrame(capomulin_df.groupby("Mouse ID")["Tumor Volume (mm3)"].mean())
# average_tumor_volume_df

cap_scat_df = pd.merge(capomulin_df, capomulin_vol_df, on="Mouse ID")
cap_scat_df = cap_scat_df.rename(columns={"Tumor Volume (mm3)_x" : "Tumor Volume (mm3)", "Tumor Volume (mm3)_y":"Average Volume"})

# scatter_plot_df
x = cap_scat_df["Weight (g)"]
y = cap_scat_df["Average Volume"]

plt.scatter(x, y)
plt.title("Capomulin Mice: Avg Tumor Vol. x Weight")
plt.xlabel("Weight (g)")
plt.ylabel("Volume (mm3)")
plt.show()
```

```python
 # Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen

#Correlation Coefficient
R = sts.pearsonr(x, y)
print(f"The correlation coefficient: {round(R[0],2)}.")

#Linear Regression
(slope, intercept, rvalue, pvalue, stderr) = sts.linregress(x, y)
regression_value = slope * x + intercept
# line_equation
line = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))

#Replot scatter plot with linear regression information
plt.scatter(x, y)
plt.plot(x, regression_value, "r-")
plt.annotate(line,(20, 37), color="red")
plt.title("Capomulin Mice: Avg Tumor Vol x Weight")
plt.xlabel("Weight (g)")
plt.ylabel("Volume (mm3)")
plt.show()
```
